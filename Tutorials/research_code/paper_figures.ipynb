```python
from stim_spike import getCache, getSessionWithMostUnits, getStimSpikePair
from analysis import getConfusionMatrix
from visualization import showConfustionMatrix

#### classifiers:
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
cache = getCache('visual-behavior-neuropixels')
session = getSessionWithMostUnits(cache)
X, cat = getStimSpikePair(session)
```

```python
conf_mat = getConfusionMatrix(X, cat, LDA())
showConfustionMatrix(conf_mat, save_path = "figures/lda_visp.png")
```

```python
area_names = ['VISl', 'VISam', 'VISpm']
for area_name in area_names:
    print (area_name)
    session = getSessionWithMostUnits(cache, area = area_name)
    X, cat = getStimSpikePair(session, area = area_name)
    conf_mat = getConfusionMatrix(X, cat, LDA())
    showConfustionMatrix(conf_mat, 4, 4, save_path = "figures/lda_" + area_name +".png")
```

```python
area_names = ['VISl', 'VISam', 'VISpm']
for area_name in area_names:
    print (area_name)
    session = getSessionWithMostUnits(cache, area = area_name)
    X, cat = getStimSpikePair(session, area = area_name, window = .1)
    conf_mat = getConfusionMatrix(X, cat, LDA())
    showConfustionMatrix(conf_mat, 4, 4, save_path = "figures/lda_100ms_" + area_name +".png")
```

```python
area_names = ['VISp', 'VISl', 'VISam', 'VISpm']
for area_name in area_names:
    print (area_name)
    session = getSessionWithMostUnits(cache, area = area_name)
    X, cat = getStimSpikePair(session, area = area_name, window = .05)
    conf_mat = getConfusionMatrix(X, cat, LDA())
    showConfustionMatrix(conf_mat, 4, 4, save_path = "figures/lda_50ms_" + area_name +".png")
```
