```python
scan_key = {'session':4, 'scan_idx':7}
trial_df = pd.DataFrame(nda.Trial & scan_key)
trial_df
```

```python

conditions = np.array([k for k in trial_df.groupby('condition_hash').groups.keys()])
np.random.shuffle(conditions)
clrs=sns.color_palette('hsv', len(conditions))
color_dict={h:c for h,c in zip(conditions,clrs)}
f, ax = plt.subplots(figsize=(12,.25))
whole_stim_line =plt.Line2D([0,40000/6.3], [0,0],color='black', linewidth=5)
ax.add_line(whole_stim_line)
for k, trial in trial_df.iterrows():
    line = plt.Line2D([trial.start_frame_time, trial.end_frame_time],
                      [0,0],
                      linewidth=5,
                      color=color_dict[trial.condition_hash])
    ax.add_line(line)

ax.set_xlim(-100,6500)
ax.set_ylim(-.5, .5)
ax.axis('off')
```
